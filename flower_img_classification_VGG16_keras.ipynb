```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.preprocessing import image
```

```python
base_dir = r"/content/drive/MyDrive/flowers"
IMAGE_SIZE = 224
BATCH_SIZE = 64
```

```python
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    validation_split = 0.1
)

validation_datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale = 1./255,
    validation_split = 0.1
)

train_datagen = train_datagen.flow_from_directory(
    base_dir,
    batch_size = BATCH_SIZE,
    target_size = (IMAGE_SIZE, IMAGE_SIZE),
    subset = 'training'
)

validation_datagen = validation_datagen.flow_from_directory(
    base_dir,
    batch_size = BATCH_SIZE,
    target_size = (IMAGE_SIZE, IMAGE_SIZE),
    subset = 'validation'
)
```

```python
from tensorflow.keras.layers import Input, Flatten, Dense
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.applications.vgg16 import VGG16
from glob import glob
```

```python
IMAGE_SIZE = [224, 224]
```

```python
vgg = VGG16(input_shape = IMAGE_SIZE+[3], weights='imagenet', include_top=False)
vgg.output
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
folders = glob.glob(r"/content/drive/MyDrive/flowers/*")
print(len(folders))
```

```python
x = Flatten()(vgg.output)
prediction = Dense(len(folders), activation='softmax')(x)
model = Model(inputs=vgg.input, outputs=prediction)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(train_datagen, steps_per_epoch=len(train_datagen), epochs=10, validation_data=validation_datagen, validation_steps=len(validation_datagen))
```

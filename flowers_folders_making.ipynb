```python
import shutil
import os
```

```python
cfolder = "/content/drive/MyDrive/flowers"
for files in os.listdir(cfolder):
  x = files.split('_')[0]
  nfolder = os.path.join(cfolder, x)
  cur = os.path.join(cfolder, files)
  new = os.path.join(nfolder, files)
  if not os.path.exists(nfolder):
    os.makedirs(nfolder)
    #print(cur, new)
    shutil.move(cur, new)
  else:
    shutil.move(cur, new)
```
